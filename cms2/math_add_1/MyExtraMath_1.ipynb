### A function that returns a polynomial of a given function:

```python
def generate_expansion(func, terms):
    """
    function that takes as arguments the expression itself in the 
    form of a string (func="sin(x)") and the number of terms of the 
    expression (terms=5). The function returns a sympy expression 
    that can then be evaluated
    """
    iteration = 0
    result = func.evalf(subs={x: 0})

    for i in range(terms):
        func = diff(func, x)
        deriv_value = func.evalf(subs={x: 0}) / factorial(iteration + 1)
        if int(deriv_value) == deriv_value:
            deriv_value = int(deriv_value)
        else:
            deriv_value
        term = deriv_value * (x ** (iteration + 1))
        result += term
        iteration += 1
    return result
func ="sin((2*x)**3)"
Taylors_polin = generate_expansion(sympify(func), terms=5)
Taylors_polin
```

## Paint function:

```python
import matplotlib.pyplot as plt
def painting(polinom, function):
    """"""
    x = symbols("x")
    xpoint = [-0.6 + i * 0.01 for i in range(100)]
    ypoint = [sympify(function).evalf(subs={x: value}) for value in xpoint]
    ypoint_polin = [polinom.evalf(subs={x: value}) for value in xpoint]

    plt.subplot(2, 1, 1)
    plt.plot(xpoint, ypoint, color = "orange")
    plt.ylabel("sin((2*x)**3) value")
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

    plt.subplot(2, 1, 2)
    plt.plot(xpoint, ypoint_polin, color = "blue")
    plt.xlabel("x value")
    plt.ylabel("sin((2*x)**3) value")
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

    plt.suptitle("Diagrams: orange - first function, blue - Taylor's function")
```

```python
#Expanded to a polynomial of the 5th degree
Taylors_polin = generate_expansion(sympify(func), terms=5)
painting(Taylors_polin, func)
```

```python
#Expanded to a polynomial of the 9th degree
Taylors_polin = generate_expansion(sympify(func), terms=9)
painting(Taylors_polin, func)
```

```python
#Expanded to a polynomial of the 15th degree
Taylors_polin = generate_expansion(sympify(func), terms=15)
painting(Taylors_polin, func)
```

```python
size = audio[0].size
time = np.arange(size)/audio[1]
dt = 1/audio[1]
```

```python
fft_ = scipy.fft.fft(audio[0])
freq = scipy.fft.fftfreq(size, dt)
```

```python
plt.plot(freq[:size//2], np.abs(fft_[:size//2]), color='red')
plt.xlabel("Frequency")
plt.title("FFT 'Robin'")
plt.grid(color="green")
plt.show()
```
